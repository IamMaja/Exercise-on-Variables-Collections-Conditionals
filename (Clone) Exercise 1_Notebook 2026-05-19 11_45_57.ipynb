### Section 3 - Logical Operators & Conditionals

```python
#COMPARISON EXPRESSIONS FOR AMARA'S RECORD
student_name = "Amara"
age = 19
subject = "Math"
score = 72
passed = True

print(score > 60)
print(age == 20)
print(subject != "Science")
print(score >= 72)
```

```python
#Logical operations
print(score >60 and passed == True)
print(age < 18 or score > 70)
print(not passed)
print(score > 80 and subject == "Math")
```

```python
#If/else statement
score = 72
if score >= 75:#75-100
    print("Distinction")
elif score >= 50:#50-74
    print("Pass")
else:#0-49:
    print("Fail")
#@45
score = 45
if score >= 75:#75-100
    print("Distinction")
elif score >= 50:#50-74
    print("Pass")
else:#0-49:
    print("Fail")
#@91
score = 91
if score >= 75:#75-100
    print("Distinction")
elif score >= 50:#50-74
    print("Pass")
else:#0-49:
    print("Fail")
```
